```python
from autogluon.tabular import TabularDataset,TabularPredictor
#训练
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_data = TabularDataset('/content/drive/MyDrive/Titanic/train.csv')
id,label = 'PassengerId','Survived'
predictor = TabularPredictor(label=label).fit(train_data.drop(columns=[id]))
```

```python
#查看训练数据
train_data.head(30)
print(train_data.shape)
```

```python
import pandas as pd
#预测
test_data = TabularDataset('/content/drive/MyDrive/Titanic/test.csv')
preds = predictor.predict(test_data.drop(columns=[id]))
submission = pd.DataFrame({id:test_data[id],label:preds})
submission.to_csv('submission.csv',index=False)
```
